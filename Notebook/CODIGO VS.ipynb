REPRESENTACION DEL GENERO MAS COMUN 

```python
genero= pd.DataFrame(master_pelis["Género"])
```

```python
genero
```

```python
genero["Género"].explode().unique()
```

```python
genero_count = genero['Género'].apply(pd.Series).stack().value_counts()
print(genero_count)
```

```python

fig = plt.figure(figsize=(11,11))

colores = ['red', 'purple', 'blue', 'orange', 'green', 'cyan', 'gold', 'gray', 'navy', 'lime', 'pink',
           'purple', 'coral', 'teal', 'yellow', 'brown', 'maroon', 'lime', 'indigo']


plt.pie(genero_count.values, labels=genero_count.index, colors=colores, autopct='%1.1f%%')
plt.axis('equal')  
plt.title('Géneros de Películas Nominadas al Oscar')
plt.show()
```
